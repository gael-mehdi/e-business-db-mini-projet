# Part 1: KNN

## Question 1

Yes, cross-validation is important when training KNN on our dataset. Indeed, the dataset consists of only 38 rows and 10 quantitative variables, with such a small dataset, there is a high risk of overfitting. Cross-validation helps assess how well the model generalizes by splitting the data into training and validation subsets multiple times.

## Question 2

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
Z = scaler.fit_transform(X)
Z
```

```python
k_values = [3, 4, 5, 6, 7]
cv = KFold(n_splits=5, shuffle=True, random_state=42) # Use KFold cross-validation

accuracy_scores = []
f1_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    accuracy = cross_val_score(knn, Z, Y, cv=cv, scoring='accuracy')
    f1 = cross_val_score(knn, Z, Y, cv=cv, scoring='f1_macro') # Use f1_macro for multi-class
    accuracy_scores.append(accuracy)
    f1_scores.append(f1)
```

```python
plt.figure(figsize=(12, 6))

# Plot accuracy scores
plt.subplot(1, 2, 1)
plt.plot(k_values, [score.mean() for score in accuracy_scores], marker='o')
plt.title('Accuracy Scores')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')

# Plot f1 scores
plt.subplot(1, 2, 2)
plt.plot(k_values, [score.mean() for score in f1_scores], marker='o')
plt.title('F1 Scores')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('F1 Score')

plt.tight_layout()
plt.show()
```

The best value of k seems to be $k=5$ , because it optimises accuracy while maintaining a good F1 score, which is crucial for evaluating performance in a multi-class problem.

## Question 3

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Train the KNN model with the optimal value of k
optimal_k = 5
knn_optimal = KNeighborsClassifier(n_neighbors=optimal_k)
knn_optimal.fit(Z, Y)

# Predict the labels
Y_pred = knn_optimal.predict(Z)

# Compute the confusion matrix
cm = confusion_matrix(Y, Y_pred)

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=knn_optimal.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

The results are very good. All the athletes are well classified except one which is predicted in class 2 but is in class 3 in reality.

## Question 4

```python
from sklearn.metrics import classification_report

# Generate the classification report
report = classification_report(Y, Y_pred, target_names=['Class 1', 'Class 2', 'Class 3'])
print(report)
```

## Question 5

```python
from sklearn.decomposition import PCA
from matplotlib.colors import ListedColormap

# Réduire les dimensions des données à 2D en utilisant PCA
pca = PCA(n_components=2)
Z_pca = pca.fit_transform(Z)

# Ajuster le modèle KNN sur les données réduites
knn_optimal.fit(Z_pca, Y)

# Créer une grille de points pour tracer les limites de décision
x_min, x_max = Z_pca[:, 0].min() - 1, Z_pca[:, 0].max() + 1
y_min, y_max = Z_pca[:, 1].min() - 1, Z_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

# Prédire les classes pour chaque point de la grille
Z_grid = np.c_[xx.ravel(), yy.ravel()]
Z_pred = knn_optimal.predict(Z_grid)
Z_pred = Z_pred.reshape(xx.shape)

# Tracer les limites de décision
plt.figure(figsize=(10, 6))
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

plt.contourf(xx, yy, Z_pred, alpha=0.3, cmap=cmap_light)
plt.scatter(Z_pca[:, 0], Z_pca[:, 1], c=Y, cmap=cmap_bold, edgecolor='k', s=20)
plt.title('Decision Boundaries of KNN (k={}) with PCA'.format(optimal_k))
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

# Part 2: SVM


## Question 1

```python
from sklearn.svm import SVC

# Define the parameter grid
C_values = [0.1, 1, 10, 100]
kernels = ['linear', 'rbf', 'sigmoid']

# Initialize lists to store the results
accuracy_scores_svm = []
f1_scores_svm = []

# Perform cross-validation for each combination of C and kernel
for C in C_values:
    for kernel in kernels:
        svm = SVC(C=C, kernel=kernel)
        accuracy = cross_val_score(svm, Z, Y, cv=cv, scoring='accuracy')
        f1 = cross_val_score(svm, Z, Y, cv=cv, scoring='f1_macro')
        accuracy_scores_svm.append((C, kernel, accuracy.mean()))
        f1_scores_svm.append((C, kernel, f1.mean()))

# Find the best combination based on accuracy
best_accuracy = max(accuracy_scores_svm, key=lambda x: x[2])
print(f"Best combination based on accuracy: C={best_accuracy[0]}, kernel={best_accuracy[1]}, accuracy={best_accuracy[2]}")

# Find the best combination based on F1-score
best_f1 = max(f1_scores_svm, key=lambda x: x[2])
print(f"Best combination based on F1-score: C={best_f1[0]}, kernel={best_f1[1]}, F1-score={best_f1[2]}")
```

```python
# Extract the accuracy and F1 scores for each combination of C and kernel
C_values = [score[0] for score in accuracy_scores_svm]
kernels = [score[1] for score in accuracy_scores_svm]
accuracy_scores = [score[2] for score in accuracy_scores_svm]
f1_scores = [score[2] for score in f1_scores_svm]

# Create a DataFrame for easier plotting
results_df = pd.DataFrame({
    'C': C_values,
    'Kernel': kernels,
    'Accuracy': accuracy_scores,
    'F1 Score': f1_scores
})

# Plot the results
plt.figure(figsize=(14, 6))

# Plot accuracy scores
plt.subplot(1, 2, 1)
for kernel in results_df['Kernel'].unique():
    subset = results_df[results_df['Kernel'] == kernel]
    plt.plot(subset['C'], subset['Accuracy'], marker='o', label=kernel)
plt.title('Accuracy Scores for SVM')
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.xscale('log')
plt.legend()

# Plot F1 scores
plt.subplot(1, 2, 2)
for kernel in results_df['Kernel'].unique():
    subset = results_df[results_df['Kernel'] == kernel]
    plt.plot(subset['C'], subset['F1 Score'], marker='o', label=kernel)
plt.title('F1 Scores for SVM')
plt.xlabel('C')
plt.ylabel('F1 Score')
plt.xscale('log')
plt.legend()

plt.tight_layout()
plt.show()
```
